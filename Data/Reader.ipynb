```python
url = "D:/2. GP/0. Data/Data/"
pdfs = os.listdir(url)
```

```python
pdfs
```

```python
import pdfplumber

data = ""

for file in os.listdir(url):
    with pdfplumber.open(url + file) as pdf:
        for page in pdf.pages:
            data += page.extract_text()
        data += '\n '
```

```python
print(data)
```

```python
with open('data.txt', 'w', encoding='utf-8') as file:
    file.write(data)
```

```python
with open('data.txt', 'r', encoding='utf-8') as file:
    read_data = file.read()
```
